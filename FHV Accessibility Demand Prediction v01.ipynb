```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor 
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df = pd.read_csv("H:/FHV Accessibility/Research Projects/Data/fhv_wav_201901-06_run-2019_08_16_.csv")
df.head()
```

```python
df.columns.values
```

```python
# Convert long and lat values to numeric
df['Pickup_longitude'] = pd.to_numeric(df['Pickup_longitude'], errors='coerce')
df['Pickup_latitude'] = pd.to_numeric(df['Pickup_latitude'], errors='coerce')

# Drop nulls
df.dropna(subset=['Pickup_longitude'], inplace=True)
df.dropna(subset=['Pickup_latitude'], inplace=True)

# Check for nulls
df.apply(lambda x: x.isnull().sum())
```

```python
features=df[['Pickup_longitude']]
labels=df[['Pickup_latitude']]

# Split dataset into training set and testing set
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size=0.5) # 50% training and 50% test
```

```python
# Random Forest
rf = RandomForestRegressor(n_estimators=500)

# Fit on training data
rf.fit(train_features, train_labels)

# Predict on test data
long_pred = rf.predict(test_features)
lat_pred = rf.predict(test_labels)

long_pred, lat_pred
```

```python
# Evaluate: mean absolute error, mean squared error, root mean squared error
# (A value of 0 indicates no error)

# mean absolute error: is the average of the absolute differences between predictions and actual values. 
# mean squared error: provides a gross idea of the magnitude of error.
# root mean squared error: the absolute value of the mean squared error.

print('Mean Absolute Error:', metrics.mean_absolute_error(test_labels, long_pred, lat_pred))
print('Mean Squared Error:', metrics.mean_squared_error(test_labels, long_pred, lat_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(test_labels, long_pred, lat_pred)))
```

```python
# Convert arrays to table
long_df = pd.DataFrame(long_pred, columns=["long"])
lat_df = pd.DataFrame(lat_pred, columns=["lat"])

new_df = pd.concat([long_df, lat_df], axis=1)
new_df
```

```python
# Export test data as csv 
header = ["long", "lat"]
new_df.to_csv('H:/FHV Accessibility/Research Projects/A Look into For Hire Vehicles Accessibility/Data/Coordinates from Random Forest Model.csv', columns=header)
```
